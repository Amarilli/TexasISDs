SAT and ACT scores merging
====

```python
    #Creating a data frame with only sat score and district name
sat_data_df.head()

```

```python
sat_math = sat_data_df.set_index(["District Name","Group"])["SAT Math"]
sat_math_cleaned = sat_math.dropna()
```

```python
sat_total = sat_data_df.set_index(["District Name","Group"])["SAT Total"]
sat_total_cleaned = sat_total.dropna()
```

```python
#Create a new DataFrame 
sat_summary = pd.DataFrame({"SAT Math": sat_math_cleaned,
                                   "SAT Total": sat_total_cleaned,})
```

```python
act_eng = act_data_df.set_index(["District Name", "Group"])["ACT English"]
act_eng_cleaned = act_eng.dropna()
```

```python
act_math = act_data_df.set_index(["District Name", "Group"])["ACT Math"].dropna()
act_reading = act_data_df.set_index(["District Name","Group"])["ACT Reading"].dropna()
act_science = act_data_df.set_index(["District Name","Group"])["ACT Science"].dropna()
act_compos = act_data_df.set_index(["District Name","Group"])["ACT Compos"].dropna()
```

```python
#Create a new DataFrame 
act_summary = pd.DataFrame({"ACT English": act_eng_cleaned,
                                   "ACT Math": act_math,
                                   "ACT Reading": act_reading,
                                   "ACT Science": act_science,
                                   "ACT Compos": act_compos
                                       })
```

```python
#Merging SAT and ACT on district 
act_sat = pd.merge(act_summary, sat_summary, how="left", on=["District Name","Group"])
act_sat
```
